# Detailed description of the main tools and interface components of Jupyter Notebook
## The Menu bar 
The menu bar in jupyter notebook provides a range of options for managing the notebook, editing cells and interacting with the kernel. 
1. File menu: Create new notebooks, open existing ones, save, and download notebooks in various formarts.
2. Edit Menu: performs common editing actions like cut, copy, paste and delete cells.
   it can also split or merge cell to organize your notebook.  
3. View Menu: Customize the jupyter Notebook interface by toggling the header, toolbar, and the line numbers.
   Access specific tools and settings 
4. Run Menu: Runs individual cells, all cells, or specific groups of cells, change cell types (Code, Markdow, Raw)
5. Kernel menu: Manage the kernel, including interrupting, restarting, or shutting it down it can also perform operations such as run all cells or clear output.
6. settings menu: this allows user to customize their experience, such customization of keyboard shortcuts, manage extensions, customization of Jupyter notebook interface and also access advance settings (e.g Code completion, syntax highlighting).
7. Help Menu: enables users to Access Jupyter Notebook documentation, keyboard shortcuts, and user interfacec tours.

   # Toolbar
   The toolbar in jupyter notebook provides quick access to common actions which include:
1. Save and checkpoint: Save your notebook and create a checkpoint, allowing you to revert to a previous version if needed
2. Insert cell: Adds a new cell above or below the current cell, giving options to insert cells in different positions.
3. Cut, Copy, Paste: Allows you to edit cells by cutting, copying and pasting content.
4. Run Cell: Executes the current cell, running the code or rendering markdown.
5. Interrupt: Stops the kernel, Clearing all variables and output.
7. Restart: Restarts the kernel, clearing all variables and output.
8. Restart Kernel and Run all cells: this action does two things:
   
 * Restart the kernel: Clears all variables, output, and state, giving you a fresh start.
   
  * Run all cells: Executes all code cells in the notebook, top-to-bottom, rendering markdown and displaying output.
   # Code cell and markdown cells
   Jupyter Notebook is a versatile tool that allows you to combine code execution and documentation seamlessly. 

1. Code Cells
Purpose: Used to write and execute code 

2. Markdown Cells
Purpose: Used for adding formatted text, headings, lists, links, images, and more to document your notebook.
Shortcut: Press Esc, then M.

## Switching Between Cell Types
Toolbar: Use the dropdown menu to select "Code" or "Markdown."
Keyboard Shortcuts:
Change to Markdown: Esc, then M.
Change to Code: Esc, then Y.
By combining these two cell types, you can create interactive, well-documented notebooks for data analysis, machine learning, or any computational task.


   # kernel
   The kernel is the computational engine that runs your code in jupyter Notebook.
   It's responsible for:
   * executing code cells,
   * managing variables and state,
   * displaying output. 
  ## How to restart the kernel 
  there are several ways to restart the kernel:
  1. Toolbar button: Click the "restart" button in the toolbar.
  2. click on the kernel button on the drop down menu
     
# Command and Edit modes 
There are two modes, Edit mode and Command mode. Edit mode allows you to type into the cells like a normal text editor. Command mode allows you to edit the notebook as a whole, but not type into individual cells.
for navigating the command mode press ESC key to enter the command mode. to switch to edit mode press the enter key.
