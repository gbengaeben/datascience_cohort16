```python
#TASK 1 
temp = float(input("Enter temperature: "))

if temp < -50 or temp > 60:
    print("Invalid")
elif temp < 10:
    print("Very Low")
elif temp <= 30:
    print("Normal")
else:
    print("High")
```

```python
#TASK 2
sales = []
for i in range(3):
    value = float(input(f"Enter sales value {i+1}: "))
    sales.append(value)

if any(v < 0 for v in sales):
    print("Invalid: Negative value found")
elif any(v > 1_000_000 for v in sales):
    print("Invalid: Value too high")
else:
    print("Sales data is valid")
```

```python
#TASK 3
age = int(input("Enter age: "))
income = float(input("Enter annual income: "))

if age < 18:
    print("Rejected")
else:
    if income > 500_000:
        print("Premium Customer")
    else:
        print("Regular Customer")
```

```python
#TASK 3
categories = {"electronics", "fashion", "food", "health"}

user_cat = input("Enter a category: ").lower()

if user_cat in categories:
    print("Valid category")
else:
    print("Invalid category")
```

```python
#TASK 5
rating = int(input("Enter rating (1–5): "))
comment = input("Enter your comment: ")
length = len(comment)

if rating < 3:
    if length > 20:
        print("Detailed Negative Review")
    else:
        print("Negative Review")
else:
    print("Positive Review")
```

```python
#TASK 6
amount = float(input("Enter transaction amount: "))
country = input("Enter transaction country: ").lower()

high_risk_countries = {"north korea", "iran", "syria", "russia"}

if amount > 500_000 or country in high_risk_countries:
    print("High Fraud Risk")
else:
    print("Safe Transaction")
```

```python
#TASK 7
columns = ["name", "age", "salary", "department"]

col = input("Enter a column name: ").lower()

if col in columns:
    print("Column exists")
else:
    print("Column missing")
```

```python
#TASK 8
score = int(input("Enter credit score (0–850): "))

if 0 <= score <= 850:
    if score <= 499:
        print("High Risk")
    elif score <= 699:
        print("Medium Risk")
    else:
        print("Low Risk")
else:
    print("Invalid Score")
```
